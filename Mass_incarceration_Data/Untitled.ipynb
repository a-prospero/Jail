### Cleaning Gender Population Data

```python
data = 'Resources/us_population_by_sex_over_time.csv'
```

```python
df = pd.read_csv(data, encoding='utf-8-sig')
```

```python
df.dropna(how='any')
```

```python
gender_pop.to_csv("Resources/gender_pop.csv",
                 encoding="utf-8", index=False, header=True)
```

# Cleaning/Merging Fed & State Admissions

```python
federal_csv = "Resources/Federal_admission_release.csv"
state_csv = "Resources/State_admission_releases.csv"
```

```python
federal_df = pd.read_csv(federal_csv, encoding='utf-8-sig')
state_df = pd.read_csv(state_csv, encoding='utf-8-sig')
```

```python
admissions_merged = pd.merge(federal_df, state_df, on="Year")
admissions_merged
```

```python
admissions_merged = admissions_merged.rename(columns={"Admissions_x": "Federal Admissions", "Releases_x": "Federal Releases",
                                             "Admissions_y": "State Admissions", "Releases_y": "State Releases"})
admissions_merged.head()
```

```python
admissions_merged.to_csv("Resources/admissions.csv",
                 encoding="utf-8", index=False, header=True)
```
